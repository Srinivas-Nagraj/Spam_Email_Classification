```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('emails.csv')
```

1.Data of 5000 Emails

```python
df['spam'].value_counts()
```

```python
x=df['text']
y=df['spam']
```

2.Train Test Split of Data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

3.Using Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(stop_words='english',lowercase=False,max_features=3000)
x_train=cv.fit_transform(x_train).toarray()
x_test=cv.transform(x_test).toarray()
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score
```

```python

accuracy_score(y_test,rf.predict(x_test))
```

```python
print(confusion_matrix(y_test,rf.predict(x_test)))
```

```python
print(precision_score(y_test,rf.predict(x_test)))
```

```python
sent1=["Srinivas,Thank you for applying to the Associate Software Engineer role at Bottomline."]
vect=cv.transform(sent1)

result=rf.predict(vect)[0]
if result==1:
    print('Spam')
else:
    print('Not Spam')
```

```python
import pickle
pickle.dump(cv,open('vectorizer.pkl','wb'))
pickle.dump(rf,open('model.pkl','wb'))
```
